```python
from tqdm import tqdm
import time

# Tạo danh sách lưu tất các title crawl được từ các web
# Danh sách ban đầu rỗng
titles = []

# Tạo danh sách chứ các url của các web cần crawl data
urls = [
    "https://kenh14.vn/doi-song.chn",
    "https://kenh14.vn/money-z.chn",
    "https://kenh14.vn/sport.chn",
    "https://kenh14.vn/xa-hoi.chn",
    "https://kenh14.vn/xa-hoi/phap-luat.chn",
    "https://kenh14.vn/the-gioi-do-day.chn",
    "https://kenh14.vn/hoc-duong.chn"
]

# Khởi tạo object webdriver
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver', options=chrome_options)

# Xét từng url trong urls theo index và data tương ứng
for id, url in enumerate(urls):

    # Tạo danh sách lưu các title crawl được ở mỗi url
    ttitles = []

    # Xuất thông tin bao gồm thứ tự, url đang crawl
    print(f"{id + 1}. Crawl URL: {url}\n")

    # Load url hiện tại vào object webdriver là biến wd với phương thức get
    wd.get(url)

    # tương tự như click nút "Xem thêm" 30 lần
    for i in tqdm(range(30), desc="Clicking"):

        # Ban đầu nút "Xem thêm" bị ẩn khỏi trang web, chỉ hiển thị khi người dùng lướt đến cuối trang
        # wd.execute_script("window.scrollTo(0, document.body.scrollHeight);") -> thực thi một đoạn code javascript nhằm lướt đến cuối thanh cuộn
        # cần làm như vậy ít nhất 3 lần để đến được cuối trang web
        for j in range(3):
            # sử dụng javascript để scroll đến cuối trang
            wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # đợi 2s sau khi scroll để web load
            time.sleep(2)

        # khi đã đên cuối trang -> gọi hàm LoadListDetail() có sẵn của web Kenh14 đê load thêm bài viết
        wd.execute_script("LoadListDetail();")
        # đợi 1s sau khi gọi hàm LoadListDetail()
        time.sleep(1)

    print('')

    # tìm các html tag thỏa mãn đường dẫn cho trước
    # "//h3[@class='knswli-title']//a" -> tìm các tag <a> chứa trong các tag <h3> có class "knswli-title"
    a_tags = wd.find_elements(By.XPATH, "//h3[@class='knswli-title']//a")

    # Vòng lặp duyệt các tag tìm được trong a_tags
    # Vòng lặp được hiển thị với process bar với tên là "Crawling"
    for i in tqdm(range(len(a_tags)), desc="Crawling"):
        try:
            # Tạo biến lưu title của mỗi tag trong a_tags
            title = a_tags[i].get_attribute('title')
        except:
            continue

        # Nếu có title thì lưu vào danh sách các title tìm được của url hiện tại
        if len(title) > 0:
            ttitles.append(title)
    
    # Dòng này để xuất ra màn hình đẹp hơn
    if id == len(urls) - 1:
        print('')

    # Xuất thông báo cho biết crawl được bao nhiu title từ mỗi url
    print(f"\nDone with {len(ttitles)} titles.\n\n")

    # Thêm các title tìm được vào danh sách tổng các title
    titles.extend(ttitles)

# Xuất thông báo cho biết tổng số lượng các title tìm được
print(f"--> Crawl total {len(titles)} titles for kenh14.vn\n")

# Tạo set các titles để lọc ra các title trùng
titles_K14 = set(titles)

# Lưu kết quả vào file 'kenh14.json' với nhãn là 0 (không châm biếm)
f = open('kenh14.txt','w')
for title in titles_K14:
  text = {"is_sarcastic": 0,"headline": title}
  f.write(str(text))
  f.write('\n')
f.close()

# Xuất thông báo cho biết tổng số lượng các title sau khi lọc các title trùng
print(f"==> Completely saved {len(titles_K14)} unique titles in kenh14.json file.")
```
